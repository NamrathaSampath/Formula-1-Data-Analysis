```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the F1 dataset from CSV
df = pd.read_csv('/Users/yuti/Downloads/merge1.csv')

# Select specific columns for correlation analysis
selected_columns = ['driverId', 'number', 'dob', 'lat', 'lng', 'lap', 'position', 'time', 'milliseconds', 'points',
                    'nationality', 'constructorId', 'forename', 'surname', 'country',
                    'fastestLap', 'fastestLapTime', 'statusId', 'positionOrder',
                    'qualifyId', 'q1', 'q2', 'q3',
                    'constructorStandingsId', 'driverStandingsId', 'year']

corr_matrix = df[selected_columns].corr()

# Set the figure size
plt.figure(figsize=(12, 10))

# Generate the correlogram using seaborn
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='Blues', square=True, annot_kws={'fontsize': 8})
plt.title('Correlation Matrix')

# Rotate the tick labels for better visibility
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get the top nationalities in descending order of counts
top_nationalities = df['nationality'].value_counts().nlargest(5).index[::-1]

# Create a sequential color palette using 'viridis' colormap
color_palette = sns.color_palette("Reds", len(top_nationalities))

# Plot the countplot with the sequential color palette and specified order
sns.countplot(data=df[df['nationality'].isin(top_nationalities)], x='nationality', palette=color_palette, order=top_nationalities)

# Customize the plot labels and title
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.title('Distribution of Top 5 Nationalities')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot for 'lap' vs 'position'
sns.scatterplot(data=df, x='lap', y='position')
plt.xlabel('Lap')
plt.ylabel('Position')
plt.title('Lap vs Position')
plt.show()
```

```python
sns.scatterplot(data=df, x='points', y='position', color='#FF0000')
plt.xlabel('Points')
plt.ylabel('Position')
plt.title('Points vs Position')
plt.show()
```
